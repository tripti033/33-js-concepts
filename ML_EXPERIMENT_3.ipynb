## ***USING SCIKIT***

```python
import numpy as np
import skimage.io as io
from skimage.transform import rescale
from skimage.color import rgb2gray

# Reading the image using its path
image = io.imread('x2.png')

# Rescale the image by a scaling factor of 0.7
rescaled_image = rescale(image, 0.7)

# Convert the image to a format supported by PIL
rescaled_image = (rescaled_image * 255).astype(np.uint8)
rescaled_image = rescaled_image[:, :, 0]

# Save the rescaled image with a filename
filename = 'savedImage.png'
io.imsave(filename, rescaled_image)

# Display the rescaled image

io.imshow(rescaled_image)
io.show()
```

## ***IMAGE AUGMENTATION***
Image augmentation is a technique used in computer vision and deep learning to artificially increase the size of a training dataset by applying various transformations to existing images. This helps improve the performance and robustness of machine learning models, especially when dealing with limited data. Common image augmentation techniques include:

Rotation: Rotate the image by a certain degree (e.g., 90 degrees) to account for variations in object orientation.

Flip: Horizontally flip the image (left to right) to account for symmetry.

Translation: Shift the image in both the x and y directions to simulate different positions of an object.

Scaling: Zoom in or out of the image to account for variations in object size.

Shearing: Apply a shearing transformation to the image to simulate tilting or slanting.

Brightness and Contrast Adjustment: Modify the brightness and contrast of the image to handle different lighting conditions.

Noise Addition: Add random noise (e.g., Gaussian noise) to the image to make it more robust to noise in real-world data.

Color Jitter: Change the color intensity or hue of the image to handle variations in color.

Crop and Resize: Crop a portion of the image and resize it to the original size, simulating different perspectives.

Elastic Distortions: Apply elastic deformations to simulate stretching or deformation.

```python
#importing all the required lib
import skimage.io as io
from skimage.transform import rotate
import numpy as np
import matplotlib.pyplot as plt
image=io.imread('x2.png')
def image_augment(image):
  fig, ax=plt.subplots(nrows=1,ncols=3,figsize=(9,8))
  ax[0].imshow(image)
  ax[0].axis('off')
  ax[1].imshow(rotate(image,angle=45,mode='wrap'))
  ax[1].axis('off')
  ax[2].imshow(np.fliplr(image))
  ax[2].axis('off')
image_augment(image)
```
